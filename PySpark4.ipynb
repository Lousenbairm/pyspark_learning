```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("DataFrame").getOrCreate()
```

```python
spark
```

```python
df_pyspark=spark.read.csv("tt3.csv", header=True, inferSchema=True)
df_pyspark.show()
```

```python
## Filter Operation
df_pyspark.filter("Salary<=300").show()
```

```python
df_pyspark.filter("Salary<=300").select(['Name', 'Age']).show()
```

```python
df_pyspark.filter(df_pyspark['Salary']<=300).show()
```

```python
df_pyspark.filter((df_pyspark['Salary']<=300) & (df_pyspark['Salary']>=200)).show()
```

```python
df_pyspark.filter(~(df_pyspark['Salary']<=300)).show()
```
